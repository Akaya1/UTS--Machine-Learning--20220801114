### **3. Jelaskan macam-macam taxonomi dalam pengerapan Machine Learning**

#### **Berikut adalah beberapa taksonomi utama yang sering digunakan:**

##### **Berdasarkan Tipe Pembelajaran**

**Supervised Learning (Pembelajaran Terawasi)**
- Deskripsi: Model dilatih menggunakan dataset yang berisi pasangan input-output yang diketahui (diberi label).
- Contoh Algoritma: Regresi linear, regresi logistik, pohon keputusan, support vector machines (SVM), k-nearest neighbor (k-NN), jaringan saraf tiruan (neural networks).
- Aplikasi: Klasifikasi email spam, prediksi harga rumah, diagnosis medis.

**Unsupervised Learning (Pembelajaran Tak Terawasi)**
- Deskripsi: Model dilatih menggunakan dataset yang tidak berlabel. Model mencoba menemukan pola atau struktur dalam data.
- Contoh Algoritma: Clustering (K-means, DBSCAN), analisis komponen utama (PCA), asosiasi (Apriori, Eclat).
- Aplikasi: Segmentasi pelanggan, pengelompokan gambar, reduksi dimensi.

**Semi-Supervised Learning (Pembelajaran Semi-Terawasi)**
- Deskripsi: Kombinasi dari supervised dan unsupervised learning, di mana model dilatih menggunakan sedikit data berlabel dan banyak data tidak berlabel.
- Contoh Algoritma: Semi-supervised SVM, graph-based methods.
- Aplikasi: Pengenalan wajah, klasifikasi teks dengan data berlabel yang terbatas.

**Reinforcement Learning (Pembelajaran Penguatan)**
- Deskripsi: Model belajar melalui interaksi dengan lingkungan dan menerima umpan balik dalam bentuk reward atau punishment.
- Contoh Algoritma: Q-learning, deep Q-network (DQN), policy gradients.
- Aplikasi: Robotika, permainan video, pengaturan strategi portofolio investasi.

##### **Berdasarkan Jenis Masalah**

**Klasifikasi**
- Deskripsi: Menetapkan input ke dalam salah satu dari beberapa kategori diskrit.
- Contoh Algoritma: Naive Bayes, decision tree, random forest.
- Aplikasi: Deteksi email spam, pengenalan wajah, diagnosa penyakit.

**Regresi**
- Deskripsi: Memetakan input ke output kontinu.
- Contoh Algoritma: Linear regression, polynomial regression, ridge regression.
- Aplikasi: Prediksi harga rumah, estimasi permintaan pasar, analisis tren ekonomi.

**Clustering**
- Deskripsi: Mengelompokkan data berdasarkan kesamaan tanpa label.
- Contoh Algoritma: K-means, hierarchical clustering, DBSCAN.
- Aplikasi: Segmentasi pasar, analisis gambar, pengelompokan dokumen.

**Dimensionality Reduction**
- Deskripsi: Mengurangi jumlah fitur dalam dataset sambil mempertahankan informasi yang paling penting.
- Contoh Algoritma: PCA, t-SNE, LDA.
- Aplikasi: Visualisasi data, prapemrosesan untuk algoritma lain, kompresi data.

##### **Berdasarkan Teknik Pembelajaran**

**Instance-Based Learning**
- Deskripsi: Model menyimpan contoh dari data pelatihan dan membuat prediksi untuk data baru berdasarkan kemiripan dengan contoh - yang disimpan.
- Contoh Algoritma: K-nearest neighbors (k-NN), case-based reasoning.
- Aplikasi: Sistem rekomendasi, klasifikasi gambar

**Model-Based Learning**
- Deskripsi: Model membuat generalisasi dari data pelatihan untuk membuat prediksi untuk data baru.
- Contoh Algoritma: Linear regression, neural networks, support vector machines (SVM).
- Aplikasi: Prediksi harga, pengenalan suara, deteksi penipuan.

##### **Berdasarkan Output**

**Discrete Output**
- Deskripsi: Model menghasilkan output yang berupa kategori atau kelas.
- Contoh Algoritma: Decision tree, Naive Bayes, random forest.
- Aplikasi: Klasifikasi dokumen, deteksi objek.

**Continuous Output**
- Deskripsi: Model menghasilkan output berupa nilai kontinu.
- Contoh Algoritma: Linear regression, ridge regression, neural networks.
- Aplikasi: Prediksi cuaca, estimasi penjualan, analisis pasar saham.

##### **Berdasarkan Metode Ensembling**

**Bagging (Bootstrap Aggregating)**
- Deskripsi: Mengkombinasikan beberapa model yang sama dengan data pelatihan yang berbeda untuk mengurangi varians.
- Contoh Algoritma: Random forest.
- Aplikasi: Klasifikasi gambar, deteksi penipuan, prediksi cuaca.

**Boosting**
- Deskripsi: Mengkombinasikan beberapa model yang lemah secara berurutan untuk mengurangi bias.
- Contoh Algoritma: AdaBoost, Gradient Boosting, XGBoost.
- Aplikasi: Pengenalan teks, deteksi penyakit, sistem rekomendasi.

**Stacking**
- Deskripsi: Menggabungkan prediksi dari beberapa model yang berbeda menggunakan model meta-learner.
- Contoh Algoritma: Stacking regressor, stacking classifier.
- Aplikasi: Prediksi rumah, pengenalan gambar, analisis sentimen.

##### **Berdasarkan Struktur Data**

**Structured Data**
- Deskripsi: Data yang terorganisir dalam format yang telah ditentukan, seperti tabel atau spreadsheet.
- Contoh Algoritma: Linear regression, decision trees, random forests.
- Aplikasi: Analisis penjualan, klasifikasi pelanggan, prediksi keuangan.

**Unstructured Data**
- Deskripsi: Data yang tidak memiliki format atau struktur yang telah ditentukan, seperti teks, gambar, atau video.
- Contoh Algoritma: Convolutional neural networks (CNN), recurrent neural networks (RNN), transformer models.
- Aplikasi: Pengenalan wajah, analisis teks, pengenalan suara.

```python
import pandas as pd

df = pd.read_csv('dataset.csv')

print(df, "\n")

# a) 
rata2_mahasiswa = df['DATANG'].mean()
rata2_mahasiswa_bulat = round(df['DATANG'].mean())
print(f"a. Rata-rata mahasiswa datang pada minggu ini: {rata2_mahasiswa} atau {rata2_mahasiswa_bulat} Mahasiswa")

# b) 
hari_dengan_biaya_tertinggi = df.loc[df['BIAYA'].idxmax()]['HARI']
print(f"b. Biaya tertinggi terjadi pada hari: {hari_dengan_biaya_tertinggi}")

# c) 
biaya_lebih_dari = df[df['BIAYA'] > 110000]['HARI'].tolist()
biaya_lebih_dari_str = ', '.join(biaya_lebih_dari)
print(f"c. Hari dengan biaya lebih dari 110000: {biaya_lebih_dari_str}")

# d) 
mahasiswa_sering_datang = df['MAHASISWA'].value_counts().idxmax()
print(f"d. Mahasiswa yang paling banyak datang ke kampus: {mahasiswa_sering_datang}")

# e) 
mahasiswa_minggu = df[df['HARI'] == 'Minggu']['MAHASISWA'].values[0]
print(f"e. Mahasiswa yang datang pada hari Minggu: {mahasiswa_minggu}")

# f) 
biaya_terbesar = df['BIAYA'].max()
biaya_terkecil = df['BIAYA'].min()
print(f"f. Biaya tertinggi: {biaya_terbesar} dan Biaya terendah: {biaya_terkecil}")

# g) 
hari_tertinggi = df['DATANG'].max()
hari_terendah = df['DATANG'].min()
print(f"g. Frekuensi datang tertinggi: {hari_tertinggi} dan Frekuensi datang terendah: {hari_terendah}")
```
